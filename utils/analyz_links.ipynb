```python
import matplotlib.pyplot as plt

# Read the file line by line
with open('/home/apohllo/src/projekty/cycloped.io/data/ru-2024/occurrences.txt', 'r') as file:
    lines = file.readlines()
```

```python
# Split each line by blank characters and count the number of tokens
line_lengths_tokens = [len(line.split()) for line in lines]

# Compute line lengths based on characters
line_lengths_chars = [len(line) for line in lines]
# Assuming 'arrays' is your collection of arrays
```

```python

tokens_filtered = [arr for arr in line_lengths_tokens if arr <= 10]
chars_filtered = [arr for arr in line_lengths_chars if arr <= 100]
```

```python
# Plot the histogram for tokens

plt.hist(tokens_filtered, bins=11, align='mid')
plt.xlabel('Number of Tokens')
plt.ylabel('Frequency')
plt.title('Histogram of Number of tokens')
plt.ticklabel_format(style='plain')
plt.show()
```

```python
# Plot the histogram for characters
plt.hist(chars_filtered, bins=50, align='mid')
plt.xlabel('Number of Characters')
plt.ylabel('Frequency')
plt.title('Histogram of Line Lengths')
plt.show()
```
